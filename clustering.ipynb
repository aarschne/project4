```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv("Resources/Aggregated_Metrics_By_Video.csv")
df2
```

```python
# Drop the header line
df2 = df2.drop(labels = 0, axis = 0)
```

```python
list(df2.columns)
```

```python
df2 = df2.drop(['Video', 'Video title', 'Video pub\xadlish time', 'Sub\xadscribers lost', 
                        'Sub\xadscribers gained', 'Av\xader\xadage per\xadcent\xadage viewed (%)', 
                        'Av\xader\xadage view dur\xada\xadtion', 'Sub\xadscribers', 
                        'Your es\xadtim\xadated rev\xaden\xadue (USD)', 'Im\xadpres\xadsions', 
                        'Im\xadpres\xadsions click-through rate (%)'], axis=1)
```

```python
df2
```

```python
list(df2.columns)
```

```python
#Fix the two columns with "\xad" in it before starting
df2.rename(columns = {'Com\xadments ad\xadded':'Comments added', 'Dis\xadlikes':'Dislikes'}, inplace=True)
```

```python
list(df2.columns)
```

```python
df2.isna().sum()
```

```python
#Since there are only two nulls, we'll just remove those two rows.
df2 = df2.dropna(axis=0)
```

```python
df2.isna().sum()
```

```python
#We are going to use Standard Scaler from scikit to standardize our data because some of it isn't in the same scale (RPM, CPM) as the others (Views, etc).
from sklearn.preprocessing import StandardScaler
scale = StandardScaler().fit_transform(df2)
df2_scaled = pd.DataFrame(scale, columns = df2.columns)
df2_scaled
```

```python
#Ok, now we've removed and fixed up my columns, we're going to visualize some of the features against each other to see if they have any potential relationship to begin with that perhaps the clustering algorithms can utilize later. We'll also utilize a heatmap to notice correlations.
import seaborn as sns
import matplotlib.pyplot as plt
corr = df2_scaled.corr()
sns.heatmap(corr, annot=True)
```

```python
df2_scaled.plot(x = "Dislikes", y = "Likes", kind="scatter")
```

```python
df2_scaled.plot(x = "Views", y = "Likes", kind="scatter")
```

```python
#There's a clear significant outlier in the data, and We're afraid it might significantly alter the data, so We're going to remove it.
df2_scaled = df2_scaled.drop(labels = 0, axis = 0)
```

```python
#Now We're going to re-run the visualizations above to see the difference.
corr = df2_scaled.corr()
sns.heatmap(corr, annot=True)
```

```python
df2_scaled.plot(x = "Dislikes", y = "Likes", kind="scatter")
```

```python
df2_scaled.plot(x = "Views", y = "Likes", kind="scatter")
```

```python
#The last step of pre-processing is use PCA to reduce the dimensionality down to 2. This will allow for better visualizations later with the clustering.
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
df2_pca = pca.fit_transform(df2_scaled)
df2_pca = pd.DataFrame(df2_pca)
df2_pca
```

```python
pca_2 = PCA(n_components = 4)
df2_pca_2 = pca_2.fit_transform(df2_scaled)
df2_pca_2 = pd.DataFrame(df2_pca_2)
df2_pca_2
```

```python
loadings = pd.DataFrame(pca_2.components_.T, columns=['PC1', 'PC2', 'PC3', 'PC4'], index = df2.columns)
loadings
```

```python
pca_2.explained_variance_ratio_
```

```python
print(pca.explained_variance_ratio_)
```

```python
#Next we'll be using both K-means clustering and agglomerative hierarcal clustering to see if this data can be clustered into distinctive groups.
from sklearn.cluster import KMeans, AgglomerativeClustering
inertia = [] 
for k in range(1, 10):
    kmeans = KMeans(n_clusters = k).fit(df2_pca_2)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), inertia, marker = "s")
plt.xlabel("$k$")
plt.ylabel("Variance") 
```

```python
#Create another KMeans model, but this time with only 3 clusters. 
kmeans = KMeans(n_clusters=3).fit(df2_pca_2)
```

```python
values = kmeans.fit_predict(df2_pca_2)
```

```python
sns.scatterplot(x = df2_pca_2[0], y = df2_pca_2[1], hue=values)
```

```python
#From the dendrogram, it appears that 3 clusters will also be the ideal number, as the height between clusters is the greatest between 3 and 4 clusters.
agg = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="ward")
```

```python
agg_values = agg.fit_predict(df2_pca_2)
```

```python
sns.scatterplot(x = df2_pca_2[0], y = df2_pca_2[1], hue=agg_values)
```

```python
#As it can be seen in the scatterplot, agglomerative clustering changed the clusters slightly. There are still two larger and more significant clusters taking up most of the data, with a few changed between them, but the smaller third cluster shrank some but is still generally the outlier plots on the data.
#After running through the two clustering algorithms, KMeans and agglomerative, they actually came out with similar results. They both used 3 clusters as the ideal number of clusters and found similarly grouped clusters. The agglomerative clustering had some minor differences but mostly kept the three clusters similar to KMeans, which was two big clusters and an smaller outlier group. 

#It's interesting there's different size clusters and We're curious to know what exactly went into deciding what video went into what cluster and what the characteristics of the three different clusters are. I would also be curious how and/or if the clusters would change with a larger sample and more diverse genre of videos. This is a good start though to understanding that there are notable differences and unseen patterns to YouTube videos simply based on a few metrics.
```
